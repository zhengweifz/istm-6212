### Question 10

How do the transaction trends over time involving pontoons compare to overall boating transaction activity?  Discuss as appropriate, adding Markdown cells for your discussion after your exploratory queries.

 I use two subqueries to contruct the final dataset. The steps are:
    1. A select statement to return the sum of numbers of overall boating activities grouped by year. I name the subquery all_boats.
    2. Another select statement to return the sum of numbers of pontoon activities grouped by year. I name this subquery pontoon.
    3. I use "left join" to join the left all_boats subquery with the right pontoon subquery. I use "left join" in case there may not be a pontoon transaction for each year.
    4. In addition to the sums of numbers of overall and pontoon activities, i also calculate pontoon to overall ratio as percentage each year. Notice that i have to multiply "pontoon.ptn_sum" by 1.0 before divsion. Otherwise sqlite would treat the divison as a integer divsion.

```python
%%sql
Select all_boats.*, pontoon.ptn_sum, (pontoon.ptn_sum * 1.0 / all_boats.all_sum) * 100 as ptn_perc
From 
    (Select year, sum(num) as all_sum From boating Group By year) all_boats 
Left Join 
    (Select year, sum(num) as ptn_sum From boating Where tx_type Like '%PONTOON%' Group By year ) pontoon 
on  all_boats.year = pontoon.year 

```

While the number of overall transaction activities were decreasing each year, the number of pontoon activities were increasing, which can be further verified by increasing pontoon activities as a percentage of overall activities each year

### Bonus (10 points)

Make a plot (inline, here, using python) that demonstrates one or more overall trends in boat registrations in Connecticut, drawing data directly from the database.

```python
#import matplotlib 
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib as mpl
import pandas as pd
import sqlite3
%matplotlib inline
mpl.style.use('ggplot')
```

```python
#grab data
sql = '''Select year, tx_type, Sum(num) as sum
         From boating
         Group By year, tx_type''';
con = sqlite3.connect('boating.db')
df = pd.read_sql(sql,con)
```

### A plot showes an overall decreasing trend

```python
#plot overall numbers
rc('figure', figsize=[6, 4])
fig = df.groupby('year').sum().plot()
fig.axes.set_xticklabels([2008, 2009, 2010, 2011, 2012, 2013, 2014])
```

```python
#turn the dataset into a wide format, using tx_type as columns
df_wide = df.pivot(index='year', columns='tx_type', values='sum')
```

```python
#increase figure size
rc('figure', figsize=[24, 40])
```

### The subplots show some transaction types, like "BOAT REG, CANOE", were decreasing, while some other transaction types, such as "RENEW PONTOON ", were increasing.

```python
#plot sum of numbers of transaction types against year to analyze trends
fig, axes = plt.subplots(7,3)
for i, col in enumerate(df_wide.columns):
    ax = axes.flat[i]
    ax.plot(df_wide.index, df_wide[col])
    ax.set_xlabel('Year')
    ax.set_ylabel(col + ' count')
    ax.set_xticklabels([2008, 2009, 2010, 2011, 2012, 2013, 2014])
```
